```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_df = pd.read_csv("titanic_train.csv.csv") 
```

```python
print(titanic_df.info())
```

```python
print(titanic_df.describe())
```

```python
print(titanic_df.isnull().sum())
```

```python
numerical_features = ['Age', 'Fare']
for feature in numerical_features:
    plt.figure(figsize=(8, 5))
    sns.histplot(data=titanic_df, x=feature, kde=True)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
categorical_features = ['Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked']
for feature in categorical_features:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=titanic_df, x=feature)
    plt.title(f'Count of {feature}')
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(titanic_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
for feature in categorical_features:
    plt.figure(figsize=(8, 5))
    sns.barplot(data=titanic_df, x=feature, y='Survived', ci=None)
    plt.title(f'Survival Rate by {feature}')
    plt.ylabel('Survival Rate')
    plt.show()
```
